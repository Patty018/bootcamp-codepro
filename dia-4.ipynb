## Challenge 3 🤺
Crear una agenda de contactos. El usuario puede elegir:

1. `Mostrar` los contactos
2. `Crear `contacto
3. `Modificar` datos del contacto

```python
agenda_de_contactos={}

def mostrar():
    print(agenda_de_contactos)

def crear():
    clave=input("Ingrese el nombre de la clave")
    valor=input("Ingrese un valor para la clave")
    agenda_de_contactos[clave]=valor


def modificar():
    clave=input("Ingrese el nombre de la clave a modificar")
    valor=input("Ingrese un nuevo valor")
    agenda_de_contactos[clave]=valor

    
    


while True:
    opcion=int(input("Ingrese una opcion: 1:Mostrar, 2:Crear, 3.Modificar , 4:Salir"))
    if opcion == 1:
        mostrar
        print(agenda_de_contactos)
    elif opcion == 2:
        crear()
        print(agenda_de_contactos)
    elif opcion == 3:
        modificar()
        print(agenda_de_contactos)
    else :
        break
```
