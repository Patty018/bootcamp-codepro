Challenge 🤺 – 10m

🎯 Objetivo: Practicar división y módulo.

Imagina que estás repartiendo caramelos: solicita al usuario ingresar el total de caramelos y la cantidad de niños, luego calcula cuántos caramelos recibe cada niño usando división entera y cuántos sobran usando módulo, e imprime ambos resultados.

```python
total_caramelos = int(input('Ingresa el total de caramelos?'))
cantidad_de_niños = int(input('Cuantos niños son?'))

caramelos_recibidos = total_caramelos // cantidad_de_niños
print(f'caramelos recibidos {caramelos_recibidos}')

caramelos_sobrantes = total_caramelos % caramelos_recibidos
print(f'sobran {caramelos_sobrantes} caramelos')
```

Challenge 🤺 - 25m

Escribir una funcion que represente un sistema de pedidos de comida rapida.

La funcion debe recibir como parametros el nombre del producto y el tamaño del pedido.

Al ejecutarse, debe imprimir en pantalla el detalle del pedido de forma clara y ordenada.

Luego, llamar a la funcion pasando los argumentos correspondientes para simular que un cliente realiza un pedido.

```python
cafe_usuario = input('Que café va a querer?')
tamaño_usuario = input('De que medida quiere su café?: pequeño, mediano o grande')

def preparar_cafe(cafe, tamaño):  #parametro
    print(f'Ya estamos preparando su café {cafe} de tamaño {tamaño}')
    print('Su café esta listo, puede retirarlo :)')
    
    
preparar_cafe(cafe_usuario, tamaño_usuario) #argumento
```

🎯 Objetivo:
Practicar variables globales y locales.

Escribir una función calcular_total que reciba el precio de un producto y calcule el total sumando un impuesto global impuesto = 10.
Usa una variable local total para almacenar el resultado e imprime el total.
Llama a la función con distintos precios para probar.

```python
impuesto = 10 

def calcular_total(total):
    print('ingrese el precio de su producto:')
    calcular_total = total + impuesto
    
calcular_total(total)

```

```python
# Variable global
impuesto = 10

def calcular_total(precio):
    # Variable local
    total = precio + impuesto
    print(f"El total a pagar es: {total}")

# Llamadas a la función con distintos precios
calcular_total(50)
calcular_total(120)
calcular_total(300)
```
