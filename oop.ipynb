## Challenge 🤺

```python
class Animal:

    def __init__(self, palabra, accion):
        self.palabra = palabra
        self.accion = accion
        
animal1 = Animal('hola', 'vueltas')
animal2 = Animal ('Cuak', 'piruetas')                         

print(f'Cuando hablo digo {animal1.palabra}')
print(f'Cuando me muevo, hago {animal1.accion}')

print(f'Cuando hablo digo {animal2.palabra}')
print(f'Cuando me muevo, hago {animal2.accion}')

    
```

## Challenge 🤺15 min
Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

Observación: Editar las clases hijas para que no se inicialicen.

```python
class Jugador:
    tipo = 'Villano'
    
    def __init__(self, nombre, capacidad):
        self.nombre = nombre
        self.capacidad = capacidad
        
    def presentarse(self):
        print(f'Soy un personaje y soy {self.tipo}')
        
    def patadas(self):
        print('Puedo dar patadas')
        
class Super_Heroe(Jugador):
    def presentarse(self):
        Super_Heroe().presentarse()
        print("También puedo salvar el mundo")
        
class persona_normal(Jugador):
    vidas = 2
    
    def pedir_ayuda(self):
        print('Necesito ser salvado ayudaaaa')
    
    
heroe = Super_Heroe("Superman", "volar")   
    
heroe.presentarse()
heroe.patadas()
heroe.volar()

persona = persona_normal("Juan", "correr")

persona.presentarse()
persona.patadas()
persona.pedir_ayuda()
```
