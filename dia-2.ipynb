Challenge 🤺 - 20m

Escribe un programa que pregunte al usuario su edad para ingresar al festival Asunciónico y determine qué tipo de acceso puede tener.

Si el usuario tiene menos de 18 imprimir: "Acceso denegado — Solo zona Kids 🎈"

Si el usuario tiene mas de 18 pero menos de 21 imprimir: "Acceso permitido — No puede comprar bebidas alcoholicas 🎟️"

Si el usuario tiene mas de 21 imprimir: "Acceso total — Disfruta el festival 🎉"

```python

# solicitamos al usuario su edad
edad_usuario = int(input('Cual es su edad?'))

#si el usuario tiene menos de 18
if edad_usuario < 18:
    print('Acceso denegado — Solo zona Kids 🎈')
    
#si el usuario tiene mas de 18 pero menos de 21
elif edad_usuario < 21:
    print('Acceso permitido — No puede comprar bebidas alcoholicas 🎟️')
    
#si el usuario tiene mas de 21
else:
    print('Acceso total — Disfruta el festival 🎉')
```

Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

Entre -10 y 13 grados, que imprima "Ro'y koko".,
Entre 14 y 23 grados, que imprima "Refresca'i".,
Entre 24 y 28 grados, que imprima "Oikoite".,
Entre 29 y 45 grados, que imprima "Hakurooooo".,
Para todo lo demás, imprimir "Ñamano mba'e!".

```python
#solicitamos al usuario que ingrese la temperatura 

temperatura = int(input('Ingrese la temperatura en grados celcius'))

#imprima la temperatura escrita
print(f'La temperatura ingresada es: {temperatura}°C')

#Entre -10 y 13 grados
if temperatura >= -10 and temperatura <= 13:
    print(f"Ro'y koko")
    
#Entre 14 y 23 grados
elif temperatura >= 14 and temperatura <= 23:
    print("Refresca'i")

#Entre 24 y 28 grados
elif temperatura >= 24 and temperatura <= 28:
    print("Oikoite")
    
#29 y 45 grados
elif temperatura >= 29 and temperatura <= 45:
    print("Hakurooooo")

 #Para todo lo demás
else:
    print("Ñamano mba'e!")

```

Challenge  - 20m

Escribir un programa que almacene una contraseña en una variable, y pregunte al usuario por la contraseña hasta que introduzca la contraseña correcta.

Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

```python
contraseña_correcta = 1234

while True:
    contraseña_ingresada = int(input('Ingrese una contraseña'))
    if contraseña_ingresada == contraseña_correcta:
        print('La contraseña es correcta')
        break
    else:
        print('La contraseña es incorrecta, intente de nuevo')
```
